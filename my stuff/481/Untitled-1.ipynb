```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
k = 8.99e9  # Coulomb's constant

# Charge location
q_location = np.array([1, 1])

# Create grid
xarray = np.linspace(-12.0, 12.0, 100)
yarray = np.linspace(-4.0, 4.0, 100)
X, Y = np.meshgrid(xarray, yarray)

# Calculate electric field components (Ex and Ey)
r = np.array([X, Y])
r0 = q_location[:, np.newaxis, np.newaxis]
r_minus_r0 = r - r0
r_norm = np.linalg.norm(r_minus_r0, axis=0)
r_norm_cubed = r_norm**3

Ex = k * (r_minus_r0[0] / r_norm_cubed)
Ey = k * (r_minus_r0[1] / r_norm_cubed)

# Create streamplot
plt.streamplot(X, Y, Ex, Ey, density=1, linewidth=1, color='k')

# Add a point representing the charge
plt.scatter(q_location[0], q_location[1], color='red', marker='o', label='Charge at (1, 1)')

# Set plot properties
plt.title('Electric Field Streamplot')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
