Interesting: There's a reason P is a change of basis matrix. For diagonailization, it converts the basis vectors to the eigenvectors of A. Thus, $Pe_1=v_1$. From the diagram above, we also know the diagonal matrix gives the eigenvalues of A.

Example: Matrix $\begin{pmatrix} 3 && 5 \\ -1 && 16 \end{pmatrix}$

$$
\begin{bmatrix} 3-\lambda && 5 \\ -1 && 16-\lambda \end{bmatrix} = 0\\
(3-\lambda)(16-\lambda)+5=0 \\
\lambda = \frac{19}{2} \pm \frac{\sqrt{149}}{2} \\
Av=\lambda v \\
\begin{pmatrix} 3 && 5 \\ -1 && 16 \end{pmatrix} \begin{pmatrix} v_1 \\ v_2 \end{pmatrix} = \begin{pmatrix} 3v_1+5v_2 \\ -1v_1+16v_2 \end{pmatrix} = \begin{pmatrix} \lambda v_1 \\ \lambda v_2 \end{pmatrix} \\
\begin{cases} 3v_1+5v_2=\lambda v_1 \\ -1v_1+16v_2 = \lambda v_2\end{cases} \\
\begin{cases} 53v_2=\lambda v_1 \\ -1v_1+16v_2 = \lambda v_2\end{cases} \\
\text{wolfram from here} \\
v = \frac{1}{2}\begin{pmatrix} 13\mp\sqrt{149}\\ 2 \end{pmatrix}
$$

From the eigenvalues and eigenvectors we can calculate the change of basis matrix and the diagonal matrix. 

$$
e_1 = \begin{pmatrix} 1 \\ 0 \end{pmatrix}, e_2 = \begin{pmatrix} 0 \\ 1 \end{pmatrix} \\
P = \frac{1}{2}\begin{pmatrix} 13-\sqrt{149} && 13+\sqrt{149}\\ 2 && 2\end{pmatrix} \\
D = \begin{pmatrix} \lambda_1 && 0\\ 0 && \lambda_2 \end{pmatrix}
$$



This is the unitary U, isn't it? Well we use the clifford unitary to diagonalize pauli strings. It has the property of converting from basis vectors to eigenvectors of the associated pauli string, hence the diagonalization thing.

By definition, a unitary has the property that it's complex conjugate is its own inverse, which is great since complex conjugates are a lot easier to find than inverse matricies.

Okay so in quantum computing we're all about diagonalization, why? It lets us change basis to z-basis for measurements, a necessity for quantum computers. AKA measurement of the paulis that make up a Hamiltonian (btw does that include $c^{[A]}$, the coefficient associated with the pauli string?)
