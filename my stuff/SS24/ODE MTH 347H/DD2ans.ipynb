$$

\frac{-1}{k}\ln(N)=t+c \\
\ln(N)=-kt+c \\
N(t) = ce^{-kt}
$$
where c encorperates  $e^c$

$N(\tau)=\dfrac{N(0)}{2}=\dfrac{c}{2}$<br>

$$

$$
inverting these we have
$$ e^{k\tau} = 2$$

$$ \ln(e^{k\tau})=\ln(2) \rightarrow k\tau=\ln(2)$$
