Lets convert to the values of a generic matrix
$$
A = \begin{bmatrix}
a && b \\ c && d
\end{bmatrix}
$$
Thus we have 

$$
A^2  =
\begin{bmatrix}
a && b \\ c && d
\end{bmatrix}
\begin{bmatrix}
a && b \\ c && d
\end{bmatrix} = 
\begin{bmatrix}
a^2 + bc & ab + a_1b \\
ac + dc & d^2 + bc
\end{bmatrix} \\

-\text{tr}(A)A = -(a+d) \begin{bmatrix} a && b \\ c && d \end{bmatrix}
= -\begin{bmatrix}
a(a + d) & (a + d)b \\
(a + d)c & d(a + d) 
\end{bmatrix} = - \begin{bmatrix}
a^2 + ad & ab + db \\
ac + dc & da + d^2
\end{bmatrix}
\\

\text{det}(A)I = \begin{bmatrix}
ad - bc & 0 \\
0 & ad - bc
\end{bmatrix}
$$

It is clear that summing these returns the empty 2x2 matrix. Thus this equality shows the Cayley-Hamilton Theorem is true
