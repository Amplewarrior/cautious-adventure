# Deep Dive #8 - Integrating Factor Method

Remember, matricies aren't actually scary. They just say that the rate of change of a variable is dependent on the current value of other variables. 

Also note $x(t)= A^{-1}x'(t)$

$e^{At} = e^t\sum_{k=0}^\infty \frac{A^k}{k!}$

Lets start by breaking everything down into its components

$$
\begin{pmatrix} x'_0(t) \\ x'_1(t) \end{pmatrix}= \begin{bmatrix} a_{11}& a_{12}\\ a_{21} & a_{22} \end{bmatrix} \begin{pmatrix} x_0(t) \\ x_1(t) \end{pmatrix} \\
$$
This gives us
$$
\begin{cases}
x'_0(t)=a_{11}x_0(t)+a_{12}x_1(t) \\
x'_1(t)=a_{21}x_0(t)+a_{22}x_1(t) \\
\end{cases}
$$

Since this is homogenous, we can split this solution into two solutions 
$$
x_a'= \begin{bmatrix} a_{11}& a_{12}\\ a_{21} & a_{22} \end{bmatrix} \begin{pmatrix} x_0(t) \\ 0 \end{pmatrix} \\
x_b'= \begin{bmatrix} a_{11}& a_{12}\\ a_{21} & a_{22} \end{bmatrix} \begin{pmatrix} 0 \\ x_1(t) \end{pmatrix} 
$$

Where $x_a'+x_b'=x'$
