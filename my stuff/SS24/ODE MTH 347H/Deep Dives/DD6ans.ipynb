We'll check if the third matrix is a linear combination of the first two

$$
\begin{bmatrix}
1 && 3 \\ 2 && -2 \\3 && -7
\end{bmatrix}
\begin{pmatrix}
a \\ b
\end{pmatrix}
=
\begin{pmatrix}
3\\ 2 \\ 1
\end{pmatrix} \\
a = \frac{3}{2},\ b = \frac{1}{2}\\
$$
As such these are linearly dependent. 
This can be verified by
$$
\begin{cases}
1*\frac{3}{2} + 3 * \frac{1}{2} = 3 \\
2 * \frac{3}{2} + 2 * \frac{1}{2} = 2\\
3 * \frac{3}{2} -7 * \frac{1}{2} = 1
\end{cases}
$$

In other words, we'll write v in the new basis





$$
v\cdot u_1 = \frac{3+2+1}{\sqrt{3}} = \frac{6}{\sqrt{3}}  \\
v\cdot u_2 = \frac{-6+2+1}{\sqrt{6}}= \frac{-3}{\sqrt{6}}  \\ 
v\cdot u_3 = \frac{0-2+1}{\sqrt{2}} = \frac{-1}{\sqrt{2}}  \\

% u_1 \cdot u_1 = \frac{3}{3}=1 \\
% u_2 \cdot u_2 = \frac{6}{6}=1 \\
% u_3 \cdot u_3 = \frac{2}{2}=1 \\
% $$

This is an orthonormal set, so we'll just use the dot products to make the expansion

$$
v = u_1\frac{6}{\sqrt{3}} + u_2  \frac{-3}{\sqrt{6}} + u_3  \frac{-1}{\sqrt{2}}
$$

Lets convert to the values of a generic matrix
$$
A = \begin{bmatrix}
a && b \\ c && d
\end{bmatrix}
$$
Thus we have 

$$
A^2  =
\begin{bmatrix}
a && b \\ c && d
\end{bmatrix}
\begin{bmatrix}
a && b \\ c && d
\end{bmatrix} = 
\begin{bmatrix}
a^2 + bc & ab + a_1b \\

\end{bmatrix} \\

-\text{tr}(A)A = -(a+d) \begin{bmatrix} a && b \\ c && d \end{bmatrix}
= -\begin{bmatrix}
a(a + d) & (a + d)b \\
(a + d)c & d(a + d) 
\end{bmatrix} = - \begin{bmatrix}
a^2 + ad & ab + db \\
ac + dc & da + d^2
\end{bmatrix}
\\

\text{det}(A)I = \begin{bmatrix}

$$
