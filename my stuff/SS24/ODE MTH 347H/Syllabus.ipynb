<h2><a name="Course_Description"></a>Course Description<strong></strong></h2>
<p>This course is an introduction to ordinary differential equations with emphasis on theory. We study different types of differential equations. We start with first order equations, finding formulas for exact solutions of separable equations and linear equations. We study qualitative techniques to understand the behavior of solutions to general first order equations and how to find approximate solutions to equations that cannot be solved exactly. The approximate solution methods include the Euler method and the Picard approximation. We use the Picard approximation to prove the existence and uniqueness of solutions to a wide class of first order nonlinear differential equations.</p>
<p>Then, we move to second order linear equations. We find exact formulas for solutions of second order differential equations with constant coefficients, with and without sources. We describe reduction of order methods to get a solution of a second order differential equation is another solution is already known. We also introduce Laplace transform methods to solve constant coefficients equations with generalized source functions including Dirac's delta generalized functions.</p>
<p>Later on we study systems of linear differential equations and the graphical representations of their solutions. We introduce systems of first order nonlinear equations, and we show how to obtain qualitative properties of their solutions by studying their linearizations around equilibrium solutions.</p>
<p>We then go back to second order scalar equations and provide a brief introduction to boundary value problems, simple Sturm-Liouville type problems, and Fourier Series. Near the end of the course, we combine previous ideas to solve an initial boundary value problem for the heat equation, which is partial differential equation.</p>

<h2><a name="Learning_Objectives"></a>Learning Objectives</h2>
<p>Students entering MTH 235 should have a firm grasp on Calculus I/II/III including differentiation and integration techniques for both single variable and multivariable functions. By the end of Calculus IV, students should know what a is differential equation, a system of differential equations, and a partial differential equation. Students should know how to find exact solutions to several types of differential equations and how to find approximate solutions to general differential equations.</p>
<p>We will learn and evaluate fundamental knowledge and skills in differential equations based on specific objectives, detailed in the list below.</p>
<h4>First Order Differential Equations</h4>
<ul>
<li>Find exact solutions to separable differential equations.</li>
<li>Find exact solutions to linear differential equations with variable coefficients.</li>
<li>Find approximate solutions to general first order differential equations using Euler method, and Picard approximation.</li>
<li>We use the Picard approximation to prove the existence and uniqueness of solutions to a wide class of first order nonlinear differential euations.</li>
</ul>
<h4>Second Order Differential Equations</h4>
<ul>
<li>Find solutions to homogeneous equations with constant coefficients.</li>
<li>Find solutions to non-homogeneous equations.</li>
<li>Use the Laplace transform to solve constant coefficients equations.</li>
<li>Understand how to describe impulsive forces with Dirac's delta generalized functions.</li>
<li>Use Laplace transform to solve differential equations with Dirac's delta generalized functions.</li>
</ul>
<h4>Systems of First Order Differential Equations</h4>
<ul>
<li>Understand what are and how to compute eigenvalues and eigenvectors of a matrix.</li>
<li>To know how to use eigenvalues and eigenvectors to find solutions of systems of first order linear differential equations.</li>
<li>Find linearizations to nonlinear systems of differential equations at equilibrium solutions.</li>
<li>Sketch and interpret phase portraits of solutions to certain nonlinear systems of differential equations.&nbsp;</li>
</ul>
<h4>Partial Differential Equations</h4>
<ul>
<li>Find solutions to two-point boundary value problems.</li>
<li>Solve simple Sturm-Liouville eigenfunction problems with Dirichlet, Neumann, and Mixed boundary conditions.</li>
<li>Compute the Fourier series expansion of piecewise continuous functions and their extensions.</li>
<li>Understand the separation of variables method to find solutions to a partial differential equations, the heat equaiton.</li>
</ul>

<h3>Homework Using Webwork</h3>
<ul>
<li>Completing the assigned homework is vital to your understanding of the subject material. You should make every effort to complete the homework assignments and seek help with problems you have not been able to solve.</li>
<li>Homework problems are similar to the problems in the course textbook. It is important that students complete the first homework assignment,<b>&nbsp;Webwork Tutorial</b>, where one can learn how to enter answers, handle mathematical symbols, simplify complicated expressions in Webwork. This assignment counts toward your total Homework Grade.</li>
<li>Check the schedule for the dates when homework sets<strong> should be done to stay on schedule</strong>. Prior to the due date at the end of the semester,&nbsp; students can try multiple times to solve each homework problem, getting immediate feedback as to whether the answer is correct or incorrect. See each homework set for the maximum number of attempts allowed. Note that clicking on the <b>Submit Answers</b>&nbsp;button at the bottom of a problem counts as an attempt, so make sure you first click on the&nbsp;<b>Preview Answers</b> button to check whether you have entered your answer in the way that you intended. Answers to homework assignments are available right after the due date, at the end of the semester.</li>
<li>There is a&nbsp;<b>Show Me Another</b>&nbsp;button that is activated after the student hits&nbsp;<b>Submit Answers</b>&nbsp;for the first time. By clicking in the&nbsp;<b>Show Me Another</b>&nbsp;button the student gets a new version of the homework problem that we call a practice problem. This practice problem has different problem constants than the original problem, and it can be solved using unlimited number of attempts.&nbsp;<b>Solving a practice problem does not count toward your homework grade</b>. Students can use the practice problem to understand how to solve such type of problems. Then the student goes back to the original homework problem and can solve it using a similar calculation as in the practice problem. The&nbsp;<b>Show Me Another</b>&nbsp;button can be used up to&nbsp;<b>five times</b>, to generate five different copies of the original homework problem.</li>
</ul>
<h3>Homework Grade</h3>
<p>Add all your homework problem grades from all homework sets. Divide that by the maximum number of points from all the homework sets. Multiply the result by 100.&nbsp;<b>This is your Homework Grade.</b></p>

<h2><a name="Course_Policies"></a>Course Policies</h2>
<h3>Face Coverings Indoors<br><strong></strong></h3>
<p>Masks are optional in all classrooms.</p>
<h3>Ignorance of the Law Excuses No One</h3>
<p>Students should read the syllabus and be familiar with our policies and how they will be graded. No make-ups or extensions will be granted for students on the basis of not knowing.</p>
<h3>Grade Change Policy</h3>
<p>If you suspect an in-person quiz or exam has been incorrectly graded please <b>return it to your instructor before leaving the in-person classroom</b> with it. Any grade revisition should be done <b>within a week</b> from the time the exam was returned to students. After this, typically <b>no grade changes can be made</b>. In addition, it is the responsibility of the student to retain all handed back material in case of any questions or concerns including grade computations.</p>
<h3>Attendance Policy</h3>
<p>Students are expected to attend all class meetings and are responsible for all of the material covered in class. Instructors teaching the class may decide to take into account attendance and/or class participation as part of the student grade. Policies to be followed will be described by each instructor in advance of their implementation. Any changes to this syllabus or in the scheduling of exams, Webwork, etc. will be announced during class meetings.</p>
<h3>Grief of Absence Policy</h3>
<p>The Mathematics faculty and staff work hard to be sensitive and to accommodate the bereavement process of a student who has lost a family member or who is experiencing emotional distress from a similar tragedy so that the student is not academically disadvantaged in their class. The Mathematics Department relies on the University's Grief Absence Policy to alert us of when it is appropriate to grant additional accommodations. According to the University's Grief Absence Policy, it is the responsibility of the student to:</p>
<ol>
<li>notify the Associate Dean or designee of their college of the need for a grief absence in a timely manner, but no later than one week from the student’s initial knowledge of the situation,</li>
<li>provide appropriate verification of the grief absence as specified by the Associate Dean, and</li>
<li>complete all missed work as determined in consultation with the instructor.</li>
</ol>
<p>It is the responsibility of the Associate Dean or designee to:</p>
<ol>
<li>determine with the student the expected period of absence – it is expected that some bereavement processes may be more extensive than others depending on individual circumstances,</li>
<li>notify the faculty that the student will be absent, and</li>
<li>receive verification of the authenticity of a grief absence request upon the student's return.</li>
</ol>
<h3>Spartan Honor Code</h3>
<p>"As a Spartan, I will strive to uphold values of the highest ethical standard. I will practice honesty in my work, foster honesty in my peers, and take pride in knowing that honor is worth more than grades. I will carry these values beyond my time as a student at Michigan State University, continuing the endeavor to build personal integrity in all that I do." More information can be found at <a href="http://honorcode.msu.edu">http://honorcode.msu.edu</a>.</p>
<h3>Academic Dishonesty</h3>
<p><strong>Academic dishonesty</strong>&nbsp;is considered a&nbsp;<strong>serious offense</strong>&nbsp;at MSU, and it also violates the&nbsp;<a href="https://honorcode.msu.edu/" target="_blank" rel="nofollow noopener">Spartan Honor Code</a>.&nbsp;<strong>Students caught cheating on an exam or quiz</strong>&nbsp;may be penalized with a&nbsp;<strong>grade reduction</strong>. An&nbsp;<strong>Academic Dishonesty Report</strong>&nbsp;describing the type of infraction will be&nbsp;<strong>sent to the Dean of the College</strong>&nbsp;in which the violating student is enrolled. Additional penalties, in accordance with MSU policy, may include the&nbsp;<strong>failure in the course</strong>&nbsp;and an&nbsp;<strong>administrative sanction</strong>, which may include&nbsp;<strong>suspension or expulsion from the University</strong>. See the&nbsp;<a href="https://www.msu.edu/~ombud/academic-integrity/index.html" rel="nofollow">MSU Academic Integrity Policies</a>.</p>
<h3>Limits of Confidentiality</h3>
<p>Essays, journals, and other materials submitted for this class are generally considered confidential pursuant to the University’s student record policies. However, students should be aware that University employees, including instructors, may not be able to maintain confidentiality when it conflicts with their responsibility to report certain issues to protect the health and safety of MSU community members and others. Instructors&nbsp; must report the following information to the Department of Police and Public Safety if you share it:</p>
<ul>
<li>Suspected child abuse/neglect, even if this maltreatment happened when you were a child,</li>
<li>Allegations of sexual assault or sexual harassment when they involve MSU students, faculty, or staff, and</li>
<li>Credible threats of harm to oneself or to others.</li>
</ul>
<p>These reports will trigger contact from the Department of Police and Public Safety who will want to talk with you about the incident that you have shared. In almost all cases, it will be your decision whether you wish to speak with that individual. If you would like to talk about these events in a more confidential setting you are encouraged to make an appointment with the MSU Counseling Center.</p>
<h3>Disruptive Behavior</h3>
<p>Article 2.III.B.4 of the Academic Freedom Report (AFR) for students at Michigan State University states: "The student's behavior in the classroom shall be conducive to the teaching and learning process for all concerned." Article 2.III.B.10 of the AFR states that "The student has a right to scholarly relationships with faculty based on mutual trust and civility." General Student Regulation 5.02 states: "No student shall . . . interfere with the functions and services of the University (for example, but not limited to, classes . . .) such that the function or service is obstructed or disrupted. Students whose conduct adversely affects the learning environment in this classroom may be subject to disciplinary action through the Student Judicial Affairs office."</p>
