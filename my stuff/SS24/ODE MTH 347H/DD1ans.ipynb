## Part 1: Scale Invariant Equations

1. Show $2tyy'-5t^2-3y^2=0$ is scale invariant for $t>0$ by finding the funtion $F$ such that $y'=F(y/t)$

$$2tyy' = 5t^2+3y^2$$
$$y' = \dfrac{5t^2+3y^2}{2ty}$$
$$=\dfrac{3y}{2t} + \dfrac{5t}{2y} = \dfrac{3}{2} \dfrac{y}{t} + \dfrac{5}{2} \dfrac{t}{y}  $$
$$\dfrac{3}{2} (y/t) + \dfrac{5}{2} (y/t)^{-1} =F(y/t)$$

2. Pf. Show the separable equation is true

$v'=\dfrac{d}{dt} \dfrac{y(t)}{t} = \dfrac{t y'(t) - y(t)}{t^2}$

$F(v)-v = F(\frac{y}{t})-\frac{y}{t}$

Substitute these into the seperable equation

$$\dfrac{v'}{F(v)-v} = \dfrac{\frac{ty'(t)-y(t)}{t^2}}{F(\frac{y}{t})-\frac{y}{t}} $$
Substitute and simplify
$$=\dfrac{ty'(t)-y(t)}{t^2(F(\frac{y}{t})-\frac{y}{t})}=\dfrac{(y'-\frac{y}{t})}{t(y'-\frac{y}{t})} = \dfrac{1}{t} $$
QED
