We want to know when U is maximized. This is at the point when the change in U is zero, which means

$$
\frac{d}{dc_1} 3/2 c_1 ^{2/3} = \frac{d}{dc_2} 3/2\delta c_2 ^{2/3} \\
c_1^{-1/3} = \delta c_2^{-1/3} \\
c_1 = \sqrt[3]{\delta} c_2 \\
c_2 = \delta^3 c_1
$$
