```python
import matplotlib.pyplot as plt

heads = []
for i in range(13):
    multiplicity = math.factorial(12) / (math.factorial(i)*math.factorial(12-i))
    heads.append(multiplicity)

plt.bar(range(len(heads)),heads)
plt.title('Results of 12 coin flips')
plt.xlabel('# of heads')
plt.ylabel('count')
plt.show()
```
