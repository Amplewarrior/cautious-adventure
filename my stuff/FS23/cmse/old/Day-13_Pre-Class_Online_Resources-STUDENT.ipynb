A critical skill for you to develop is the ability to find and use new coding tools from online sources. If you want to learn these tools on your own, it’s best to use a coding skill that we'll refer to as **tinkering.** 


## Code tinkering

Tinkering is where you experiment with a functioning piece of code that is new to you to understand it well enough to use it for your purposes. 

**Note:** Tinkering is not an official term; it’s just what we’re calling this kind of code experimentation. 


### Important Ideas for Tinkering with Code

1. **Always have a copy of the original code, in case you break things too much and need to start over.** You can use the copy/comment/paste method (not officially recognized as a thing; just what we’re calling it). 
1. **Make sure there is some output to see the effect of the changes you make.** Either plot the results or print them, with a preference for plotting (easier to see changes!) 
1. **Look for specific values in the code that you can change.** Change the value of any numbers/strings you see. 
1. **Test your understanding of the code.** When you feel like you understand how things work, make some predictions about what will happen if you change specific values.
1. **Use your understanding of the code to get it to work for your purposes.** Ultimately, we tinker with code to figure out how to use new tools that we find online.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#import the tables
gini = pd.read_csv('cia-gini.csv', delimiter=';')
unemployment = pd.read_csv('cia-unemployment.csv', delimiter=';')
life_expecancy = pd.read_csv('cia-life_expectancy.csv', delimiter=';')

#reorder rows based on gini
nations = gini['name']
unemployment = unemployment.set_index('name')
unemployment = unemployment.reindex(nations)

life_expecancy = life_expecancy.set_index('name')
life_expecancy = life_expecancy.reindex(nations)

gini_lim = gini['value'].min(), gini['value'].max()
expectancy_lim = life_expecancy['value'].min(), life_expecancy['value'].max()
unemployment_lim = unemployment['value'].max(), unemployment['value'].max()

#matplot time
fig, (ax0, ax1) = plt.subplots(ncols=2, figsize=(9, 4))

hb = ax0.hexbin(gini['value'], unemployment['value'], gridsize=20, cmap='inferno')
ax0.xlim=gini_lim
ax0.ylim=unemployment_lim
ax0.set_title("Gini Index vs Life Expectancy")
cb = fig.colorbar(hb, ax=ax0)
ax0.set_xlabel('Gini Index')
ax0.set_ylabel('Unemployment (%)')
#ax0.set_yscale('log')


hb = ax1.hexbin(gini['value'], life_expecancy['value'], gridsize=20, cmap='inferno')
ax1.xlim=gini_lim
ax1.ylim=expectancy_lim
ax1.set_title("Gini Index vs Unemployment")
cb = fig.colorbar(hb, ax=ax1)
ax1.set_xlabel('Gini Index')
ax1.set_ylabel('Life Expectancy at Birth (Years)')

plt.show()

### unused code
#get the right rows
#location_list = gini['name'].to_numpy()
#unemployment_vals = unemployment['value'].to_numpy()
#life_expecancy_vals = life_expecancy['value'].to_numpy()
#
#link up data using numpy
```
